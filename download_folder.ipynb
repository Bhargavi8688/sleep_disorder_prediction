# Download Sleep Disorder Project as ZIP File

This notebook will **zip your entire project folder** and let you download it.

### Steps:
1. Run each cell in order
2. After running cell 3, a file called `sleep_disorder_project.zip` will be created
3. Right-click the zip file in the VS Code Explorer panel → **Download**

## Step 1: Import Required Libraries

```python
import os
import zipfile
import shutil
import pandas as pd
from IPython.display import FileLink, display

print("Libraries imported successfully!")
```

## Step 2: Create ZIP File of the Project Folder

```python
# Define the folder to zip and the output zip file name
project_folder = "/workspaces/sleep_disorder"
zip_output_path = "/workspaces/sleep_disorder/sleep_disorder_project.zip"

# Folders/files to exclude from the zip
exclude_dirs = {'.git', 'node_modules', '__pycache__', '.gradle', 'build'}

# Create the ZIP file
with zipfile.ZipFile(zip_output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(project_folder):
        # Skip excluded directories
        dirs[:] = [d for d in dirs if d not in exclude_dirs]
        for file in files:
            file_path = os.path.join(root, file)
            # Skip the zip file itself
            if file_path == zip_output_path:
                continue
            arcname = os.path.relpath(file_path, project_folder)
            zipf.write(file_path, arcname)

zip_size_mb = os.path.getsize(zip_output_path) / (1024 * 1024)
print(f"ZIP file created: {zip_output_path}")
print(f"ZIP file size: {zip_size_mb:.2f} MB")
```

## Step 3: List Files Inside the ZIP

```python
# List all files packed inside the ZIP
with zipfile.ZipFile(zip_output_path, 'r') as zipf:
    file_list = zipf.namelist()

print(f"Total files in ZIP: {len(file_list)}")
print("\nFiles included:")
for f in file_list[:30]:  # Show first 30 files
    print(f"  {f}")
if len(file_list) > 30:
    print(f"  ... and {len(file_list) - 30} more files")
```
